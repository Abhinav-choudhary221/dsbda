```python
import pandas as pd
import numpy as np
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = """Iris dataset is widely used in machine learning.
It contains sepal and petal measurements of flowers."""
print(text)
```

```python
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

```python
pos_tags = nltk.pos_tag(tokens)
print("POS Tags:\n", pos_tags)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("After Stopword Removal:\n", filtered_tokens)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized Words:\n", lemmatized_words)
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.info()
```

```python
df['text'] = df.astype(str).agg(' '.join, axis=1)

df[['text']].head()
```

```python
documents = df['text'].tolist()

print(documents[:3])
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
print(feature_names)
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

tfidf_df.head()
```
